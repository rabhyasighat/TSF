```python
state = df.groupby("State")[["Sales","Profit"]].sum().sort_values(by = "Sales", ascending = False)
plt.figure(figsize = (20,20))
state[:25].plot(kind = "bar",color = ["k","yellow"],edgecolor = "#000000")
plt.title("Profit Or Loss & Sales of top 25 States")
plt.xlabel("States")
plt.ylabel("Total profit / loss and Sales")
plt.grid(True)
state[25:].plot(kind = "bar",color = ["k","yellow"],edgecolor = "#000000")
plt.title("Profit Or Loss & Sales of the least economic States")
plt.xlabel("States")
plt.ylabel("Total profit / loss and Sales")
plt.grid(True)
```

```python
df.groupby('Category')['Profit','Sales'].sum().plot.bar(color = ['r','k'])
df.groupby('Ship Mode')['Profit','Sales'].sum().plot.bar(color = ['g','k'])
df.groupby('Segment')['Profit','Sales'].sum().plot.bar(color = ['pink','k'])
df.groupby('Sub-Category')['Profit','Sales'].sum().plot.bar(color = ['b','k'])
```

```python
sub = df.groupby("Sub-Category")["Quantity"].sum().sort_values(ascending = False).to_frame()
plt.figure(figsize = (10,6))
sns.barplot(x = sub.index,y = sub["Quantity"],label = "Quantity", palette = 'ch:r=.2')
plt.title(" Number of Products Sold")
plt.xlabel("Product")
plt.ylabel("Total Quantities sold")
plt.legend()
plt.xticks(rotation = "vertical")
plt.grid(True)
plt.show()
```

```python
sub1 = df.groupby("Sub-Category").sum()['Profit'].sort_values(ascending=True)
sub1.plot.bar(figsize=(12,5), color = 'lightblue')
plt.show()
```

<h5>From this graph we decipher that some products (tables, supplies.. etc) plotted below 0 are in loss. Whereas, the other products (copies, phones.. etc), plotted above 0, are adding to the profit of the company.</h5>

```python
 pd.DataFrame(df.groupby('State').sum())['Profit'].sort_values(ascending=True)
```
