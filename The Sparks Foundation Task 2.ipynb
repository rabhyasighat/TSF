<h1>K-MEANS MODELLING</h1>

finding the number of clusters for K-means clusttering

```python
z = iris.iloc[:, [0, 1, 2, 3]].values
z
```

```python
from sklearn.cluster import KMeans
wcss = []
for k in range(1, 11):
    # k-means++ is to avoid the poor clusterings found by the standard k-means algorithm
    kmeans = KMeans(n_clusters = k, random_state = 0)
    kmeans.fit(z)
    wcss.append(kmeans.inertia_)
```

```python
#plotting the elbow graph
plt.figure(figsize = (8,5))
plt.plot(range(1, 11), wcss, color = 'red', marker = '^')
plt.title('The elbow method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS') # sum of squares within clusters
plt.show()
```

<h5><i>The number of clusters inferred is "3"</i></h5>

```python
#applying kmeans to the dataset
kmeans = KMeans(n_clusters = 3, random_state = 0)
y_means= kmeans.fit_predict(z)
```

```python
kmeans.labels_
```

```python
plt.figure(figsize = (10,5))
plt.scatter(z[y_means == 0, 0], z[y_means == 0, 1], 
            s = 100, c = 'k', label = 'Iris-setosa')
plt.scatter(z[y_means == 1, 0], z[y_means == 1, 1], 
            s = 100, c = 'red', label = 'Iris-versicolour')
plt.scatter(z[y_means == 2, 0], z[y_means == 2, 1],
            s = 100, c = 'yellow', label = 'Iris-virginica')

# Plotting the centroids of the clusters
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:,1], 
            s = 100,  color = 'blue', label = 'Centroids')

plt.legend(bbox_to_anchor = (1,1), loc = 2)
```
